## 1- Import of libraries, loading and general visualisation of the data

## 2- Filtrado de datos

```python
#Data filtering to remove erroneous and disposable data

df = df[df["Gender"].isin(["Male", "Female"])]
df = df[df["Major_Drawback"].isin(["Yes, issues with product quality", "No, I have not experienced any drawbacks", "Yes, delayed or problematic deliveries", "Yes, difficulty with returns or refunds"])]
```

## 3- Representacin grfica de los datos

```python
#Delete in a copy the column that is not relevant for the graph to be created
df_bar = df.drop(["Desired_Improvements"], axis=1)
```

```python
#Display of the columns of the dataset minus Desired_Improvements in pie charts

plt.figure(figsize=(6,4))
for col in df_bar:
    df[col].value_counts().plot.pie(
    autopct='%1.1f%%',
    shadow=True)
    plt.title(f"Distribution according to {col}")
    plt.ylabel("")
    plt.show()
```

## Gender-relevant line graph visualisation of relevant column columns 

```python
# Group data by "Online_Shopping_Freq" and "Gender" and count occurrences
grouped = df.groupby(['Online_Shopping_Freq', 'Gender']).size().reset_index(name='Cantidad')

# Creation of the chart
plt.figure(figsize=(8, 5))

# Iteration on each gender 
for gender in grouped['Gender'].unique():
    subset = grouped[grouped['Gender'] == gender]
    plt.plot(subset['Online_Shopping_Freq'], subset['Cantidad'], label=gender, marker='o')

# Customisation of the chart
plt.xlabel("Frequency of online shopping")
plt.ylabel("Total")
plt.title("Frequency of online shopping by Gender")
plt.legend(title="Gender")  
plt.grid(True)  
plt.tight_layout()  

plt.show()
```

```python
# Group data by "Attraction_Factor" and Gender and count occurrences
grouped = df.groupby(['Attraction_Factor', 'Gender']).size().reset_index(name='Cantidad')

# Creation of the chart
plt.figure(figsize=(8, 5))

# Iteration on each gender 
for gender in grouped['Gender'].unique():
    subset = grouped[grouped['Gender'] == gender]
    plt.plot(subset['Attraction_Factor'], subset['Cantidad'], label=gender, marker='o')

# Customisation of the chart
plt.xlabel("Attractive factors")
plt.ylabel("Total")
plt.title("Attractive factors by gender")
plt.legend(title="Gender")  
plt.grid(True)  
plt.tight_layout()  
plt.xticks(rotation=90) 

plt.show()
```

```python
# Group data by "Comfortable_Price_Range" and Gender and count occurrences
grouped = df.groupby(["Comfortable_Price_Range", "Gender"]).size().reset_index(name='Cantidad')

# Creation of the chart
plt.figure(figsize=(8, 5))

# Iteration on each gender 
for gender in grouped["Gender"].unique():
    subset = grouped[grouped["Gender"] == gender]
    plt.plot(subset["Comfortable_Price_Range"], subset["Cantidad"], label=gender, marker="o")

# Customisation of the chart
plt.xlabel("Comfortable price range")
plt.ylabel("Total")
plt.title("Comfortable price range by gender")
plt.legend(title="Gender")  
plt.grid(True)  
plt.tight_layout()  

plt.show()
```

```python
# Group data by "Frequent_Products" and Gender and count occurrences
grouped = (
    df.groupby(["Frequent_Products", "Gender"])
    .size()
    .unstack(fill_value=0)
    .stack()
    .reset_index(name="Cantidad")
)

# Creation of the chart
plt.figure(figsize=(8, 5))
for gender, subset in grouped.groupby("Gender"):
    plt.plot(subset['Frequent_Products'], subset['Cantidad'], label=gender, marker='o')

# Customisation of the chart
plt.xlabel("Frequent products")
plt.ylabel("Total")
plt.title("Frequent products bought by gender")
plt.legend(title="Gender")
plt.grid(True)
plt.tight_layout()
plt.xticks(rotation=90)

plt.show()
```

```python
# Creation of a word cloud to visualise the most repeated and relevant desired improvements
# All desired improvements are merged into one string
texto = ' '.join(df["Desired_Improvements"])

# Creation of the word cloud
wordcloud = WordCloud(
    width=800,              # Ancho de la imagen
    height=400,             # Alto de la imagen
    background_color="white",  # Color de fondo
    max_words=50,           # Nmero mximo de palabras
    colormap="viridis"      # Mapa de colores (opcional)
).generate(texto)

# Represent the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")  
plt.title("Word cloud of desired improvements", fontsize=16,y=1)
plt.show()
```
